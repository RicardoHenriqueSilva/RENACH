```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar os dados
file_path = r'C:\Users\ricardo\Desktop\RENACH\DADOS TRATADOS\condutores-habilitados-fev-2024-tratado.xlsx'
dados_condutores = pd.read_excel(file_path)

# Obter as colunas que não são a quantidade de condutores históricos
colunas_x = dados_condutores.columns[dados_condutores.columns != 'Qt. Condutor Histórico']

# Criar gráficos de barras para cada categoria
for coluna_x in colunas_x:
    plt.figure(figsize=(10, 6))
    sns.barplot(x=coluna_x, y="Qt. Condutor Histórico", data=dados_condutores)
    plt.title(f"Relação da quantidade de condutores históricos por {coluna_x}")
    plt.xticks(rotation=90)
    plt.show()
```
